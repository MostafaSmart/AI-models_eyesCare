```python
import os
import pandas as pd
from PIL import Image

image_folder = "/content/DDR-dataset/train"
csv_file = "/content/trainLabels.csv"

# المجلدات المستهدفة بناءً على التصنيفات
target_dirs = {
    0: "data/No_DR",
    1: "data/Mild",
    2: "data/Moderate",
    3: "data/Severe",
    4: "data/Proliferative_DR"
}

# الامتدادات المدعومة
supported_extensions = ['jpeg', 'jpg', 'png']

# قراءة ملف CSV
df = pd.read_csv(csv_file)

# إنشاء المجلدات إذا لم تكن موجودة
for target_dir in target_dirs.values():
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

# معالجة الصفوف بحيث يتم الاحتفاظ بصور واحدة فقط عند تكرار ID بنفس التصنيف 0
unique_rows = []
i = 0
while i < len(df):
    current_row = df.iloc[i]
    image_name = current_row['image']
    level = current_row['level']

    # تحقق من السطر التالي
    if i + 1 < len(df):
        next_row = df.iloc[i + 1]
        next_image_name = next_row['image']
        next_level = next_row['level']

        # إذا كانت الصورة التالية لها نفس الـ ID ونفس التصنيف 0
        if image_name.split('_')[0] == next_image_name.split('_')[0] and level == next_level == 0:
            unique_rows.append(current_row)  # احتفظ بالسطر الأول
            i += 2  # تخطى السطر التالي لأنه مكرر
        else:
            unique_rows.append(current_row)
            i += 1
    else:
        unique_rows.append(current_row)
        i += 1

# تحويل إلى DataFrame جديد
unique_df = pd.DataFrame(unique_rows)

# إجمالي عدد الصور
total_images = len(unique_df)
processed_count = 0  # عداد الصور التي تم معالجتها

# المرور على البيانات في DataFrame الجديد
for index, row in unique_df.iterrows():
    image_name = row['image']
    level = row['level']

    # البحث عن الصورة بأي امتداد مدعوم
    image_path = None
    for ext in supported_extensions:
        temp_path = os.path.join(image_folder, f"{image_name}.{ext}")
        if os.path.exists(temp_path):
            image_path = temp_path
            break

    # إذا تم العثور على الصورة
    if image_path:
        # فتح الصورة وتغيير حجمها
        img = Image.open(image_path)
        img_resized = img.resize((640, 640))

        # مسار حفظ الصورة في المجلد المستهدف مع تحويلها إلى صيغة JPG
        save_path = os.path.join(target_dirs[level], f"{image_name}.jpg")

        # حفظ الصورة بصيغة JPG
        img_resized = img_resized.convert("RGB")  # التأكد من التحويل إلى RGB قبل الحفظ بصيغة JPG
        img_resized.save(save_path, "JPEG")

        # تحديث العداد وطباعة التقدم
        processed_count += 1
        print(f"Processed {processed_count}/{total_images} images")
    else:
        print(f"الصورة {image_name} غير موجودة بأي امتداد مدعوم.")
```

```python
import os
import random
from PIL import Image

def resize_and_convert_images(src_folder, dest_folder, num_images=100, size=(640, 640)):
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    # الحصول على اسم المجلد
    folder_name = os.path.basename(src_folder)

    # الحصول على قائمة الصور في مجلد المصدر
    all_files = [f for f in os.listdir(src_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    # اختيار 100 صورة عشوائية
    selected_files = random.sample(all_files, num_images)

    for file_name in selected_files:
        # فتح الصورة
        file_path = os.path.join(src_folder, file_name)
        img = Image.open(file_path)

        # تغيير حجم الصورة
        img = img.resize(size, Image.ANTIALIAS)

        # تحويل الصورة إلى JPEG إذا لم تكن بالفعل بتلك الصيغة
        if img.format != 'JPEG':
            file_name = os.path.splitext(file_name)[0] + '.jpg'

        # تعديل اسم الصورة بإضافة اسم المجلد
        base_name = os.path.splitext(file_name)[0]
        new_file_name = f'{folder_name}_{base_name}.jpg'

        # حفظ الصورة في المجلد الهدف
        dest_path = os.path.join(dest_folder, new_file_name)
        img.save(dest_path, 'JPEG')


```

```python
# تحديد مسارات المجلدات
src_folder = '/content/data2/dataset/Mild'
dest_folder = 'FinalData/Mild'

# استدعاء الدالة
resize_and_convert_images(src_folder, dest_folder)
```

```python
import os
import random
import shutil

def copy_random_images(src_folder, dest_folder, num_images=708):
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    # الحصول على قائمة الصور في مجلد المصدر
    all_files = [f for f in os.listdir(src_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    # اختيار 300 صورة عشوائية
    selected_files = random.sample(all_files, num_images)

    for file_name in selected_files:
        # تحديد مسارات الملفات
        src_path = os.path.join(src_folder, file_name)
        dest_path = os.path.join(dest_folder, file_name)

        # نسخ الصورة إلى المجلد الهدف
        shutil.copy(src_path, dest_path)

```

```python
import os
import shutil
import random

def split_data(src_folder, dest_folder, train_ratio=0.8, test_ratio=0.1, valid_ratio=0.1):
    # التأكد من إنشاء المجلدات الأساسية
    for subfolder in ['train', 'test', 'validation']:
        subfolder_path = os.path.join(dest_folder, subfolder)
        if not os.path.exists(subfolder_path):
            os.makedirs(subfolder_path)

    # الحصول على أسماء التصنيفات (المجلدات الفرعية)
    categories = [d for d in os.listdir(src_folder) if os.path.isdir(os.path.join(src_folder, d))]

    for category in categories:
        category_path = os.path.join(src_folder, category)

        # التأكد من إنشاء مجلدات التصنيف
        for subfolder in ['train', 'test', 'validation']:
            dest_category_path = os.path.join(dest_folder, subfolder, category)
            if not os.path.exists(dest_category_path):
                os.makedirs(dest_category_path)

        # الحصول على قائمة الصور في التصنيف
        all_files = [f for f in os.listdir(category_path) if os.path.isfile(os.path.join(category_path, f))]
        random.shuffle(all_files)  # خلط قائمة الصور عشوائيًا

        # حساب أعداد الصور لكل مجموعة
        total_files = len(all_files)
        num_train = int(total_files * train_ratio)
        num_test = int(total_files * test_ratio)
        num_valid = total_files - num_train - num_test  # لضمان التوزيع الصحيح

        # تقسيم الصور
        train_files = all_files[:num_train]
        test_files = all_files[num_train:num_train + num_test]
        valid_files = all_files[num_train + num_test:]

        # نقل الصور إلى المجلدات المناسبة
        for file_name in train_files:
            shutil.copy(os.path.join(category_path, file_name), os.path.join(dest_folder, 'train', category, file_name))
        for file_name in test_files:
            shutil.copy(os.path.join(category_path, file_name), os.path.join(dest_folder, 'test', category, file_name))
        for file_name in valid_files:
            shutil.copy(os.path.join(category_path, file_name), os.path.join(dest_folder, 'validation', category, file_name))

```

```python

# تحديد مسارات المجلدات
src_folder = '/content/data111/dataset'
dest_folder = '/content/dataw112233'

# استدعاء الدالة
split_data(src_folder, dest_folder)
```
